# Sub-Gaussian Proxy Variance Examples of use

This notebook demonstrates how to use different distributions and their sub-Gaussian variance proxies.

```python
import sys
import os

# Add the project root directory to Python path
project_root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
if project_root not in sys.path:
    sys.path.append(project_root)

from src.variance_proxy import *
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Asymmetric 3-Mass Distribution

We'll demonstrate both the easy and hard regimes.

```python
# Hard Regime Example (p3 > 4*sqrt(p1*p2))
p1, p2 = 0.15, 0.25
p3 = 1 - p1 - p2
print(f"Easy Regime? {p3 <=  4*np.sqrt(p1*p2)}")

proxy_easy = SubGaussian3MassAsymmetricProxy(p1, p2, a=1)
sigma2_opt, lambda_star = proxy_easy.subgaussian_optimal_variance_proxy()
print(f"\nTrue variance: {proxy_easy.variance:.6f}")
print(f"Optimal proxy variance: {sigma2_opt:.6f}")
print(f"Optimal lambda: {lambda_star:.6f}")

# Plot will work because we're in hard regime
proxy_easy.plot_objective_function()
```

## 2. Symmetric 3-Mass Distribution

```python
p = 0.15  # probability at each tail
a = 1.0   # scale parameter

proxy_sym = SubGaussian3MassSymmetricProxy(p, a)
sigma2_opt, lambda_star = proxy_sym.subgaussian_optimal_variance_proxy()
print(f"True variance: {proxy_sym.variance:.6f}")
print(f"Optimal proxy variance: {sigma2_opt:.6f}")
print(f"Optimal lambda: {lambda_star:.6f}")
proxy_sym.plot_objective_function()
```

## 3. Triangular Distribution

```python
# Asymmetric triangular distribution
a, b = 1.5, 7
proxy_tri = SubGaussianTriangularProxy(a, b)
res = proxy_tri.subgaussian_optimal_variance_proxy()

print(f"True variance: {proxy_tri.variance:.6f}")
print(f"Optimal proxy variance: {res['sigma_opt_squared']:.6f}")
print(f"Optimal lambda: {res['lambda_opt']:.6f}")

proxy_tri.plot_objective_function()
```

## 4. Beta Distribution

```python
# Asymmetric Beta distribution
alpha, beta = 1, 3
proxy_beta = SubGaussianBetaProxy(alpha, beta)
sigma2_opt, lambda_star = proxy_beta.subgaussian_optimal_variance_proxy()

print(f"True variance: {proxy_beta.variance:.6f}")
print(f"Optimal proxy variance: {sigma2_opt:.6f}")
print(f"Optimal lambda: {lambda_star:.6f}")

proxy_beta.plot_objective_function()
```
